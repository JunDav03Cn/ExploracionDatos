```python
import pandas as pd

url = "https://github.com/javierherrera1996/lecture_analytics/raw/refs/heads/main/datasets/exploracion_datos/game.csv"

df = pd.read_csv(url)

print(df.head())
```

```python
filas, columnas = df.shape

print(f"El dataset tiene {filas} registros (filas) y {columnas} variables (columnas).")
```

```python
print(df.info())

tipos = df.dtypes.value_counts()
print("\nDistribución de tipos de variables:")
print(tipos)

print("\nColumnas por tipo:")
for tipo, columnas in df.dtypes.groupby(df.dtypes).groups.items():
    print(f"{tipo}: {list(columnas)}")
```

```python
valores_faltantes = df.isnull().sum()
print("Valores faltantes por columna:")
print(valores_faltantes)

print(f"\nTotal de valores faltantes: {df.isnull().sum().sum()}")
```

```python
df = df.dropna()
```

```python
print("Columnas del dataset:")
print(df.columns)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['plus_minus_home'], kde=True)
plt.title("Distribución de plus_minus_home")
plt.show()
```

```python
sns.boxplot(x=df['plus_minus_home'])
plt.title("Boxplot de plus_minus_home")
plt.show()
```

```python
sns.scatterplot(x=df['pts_away'], y=df['plus_minus_home'])
plt.title("Relación entre pts_away y plus_minus_home ")
plt.show()
```

```python
df_numerico = df.select_dtypes(include=['number'])

correlacion = df_numerico.corr()

print("Matriz de correlación:")
print(correlacion)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor de correlación")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['season_id'])
plt.title("Boxplot de la season_id")
plt.show()
```

```python
from scipy.stats import zscore

z_scores = zscore(df.select_dtypes(include=['number']))

outliers = (abs(z_scores) > 3)

outliers_df = df[outliers.any(axis=1)]
print("Registros atípicos:")
print(outliers_df)
```

```python
df_cleaned = df[~outliers.any(axis=1)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x='team_abbreviation_home', y='pts_away', data=df)
plt.title("Comparación entre categorías")
plt.xlabel("team_abbreviation_home")
plt.ylabel("pts_away")
plt.show()
```

```python
sns.lineplot(x='min', y='pts_away', data=df)
plt.title("Evolución de los pts_away a lo largo del tiempo")
plt.xlabel("min")
plt.ylabel("pts_away")
plt.show()
```

```python
sns.violinplot(x='team_abbreviation_home', y='pts_away', data=df)
plt.title("Distribución de pts_away por team_abbreviation_home")
plt.show()
```
